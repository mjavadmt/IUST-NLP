```python
from bs4 import BeautifulSoup
import requests
url = "https://meta.stackoverflow.com/questions/421831/temporary-policy-chatgpt-is-banned#:~:text=ChatGPT%20is%20an%20Artificial%20Intelligence,rules"
req = requests.get(url)
soup = BeautifulSoup(req.text, "html.parser")
get_answers = soup.find_all("div", {"class": "s-prose js-post-body"})
for element in get_answers:
    find_p_tag = element.find_all("p")
    first_p_tag = find_p_tag[0].text
    print(first_p_tag[:50])
    print("--" * 50) 
```
