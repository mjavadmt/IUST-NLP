```python
import gensim.downloader as api
wv_from_bin = api.load("glove-wiki-gigaword-200")
```

### <font color="red"> Using analogies, go from 'hospital' to 'ambulance' </font>

#### Hint: King - Man + Woman = Queen

```python
print(wv_from_bin.most_similar(positive=['ambulance', 'school'], negative=['hospital']))
```

```python
print(wv_from_bin.most_similar(positive=['ambulance', 'airport'], negative=['hospital']))
```

```python
print(wv_from_bin.most_similar(positive=['ambulance', 'library'], negative=['hospital']))
```

```python
print(wv_from_bin.most_similar(positive=['ambulance', 'station'], negative=['hospital']))
```

## as we can see hospital has ambulance so (library has book), (school has students), (airport has plane), (station has bus-train)

### <font color="red"> Using analogies, find some gender biases learned by model from corpus regarding 'toys'</font>
